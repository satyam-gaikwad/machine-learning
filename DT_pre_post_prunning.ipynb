```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

x_tr,x_te,y_tr,y_te = train_test_split(iris.data,iris.target)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## postprunning
treeM = DecisionTreeClassifier(max_depth=2)
```

```python
treeM.fit(x_tr,y_tr)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeM,filled=True)
```

```python
y_pred = treeM.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,y_te))
```

## Pre-prunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeM = DecisionTreeClassifier()
```

```python
parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

### hyperparameter tunning

```python
cv = GridSearchCV(treeM,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(x_tr,y_tr)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_pred,y_te))
```

```python
print(classification_report(y_pred,y_te))
```

```python

from sklearn import tree
tree.plot_tree(cv.best_estimator_)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
# Load the dataset
df = load_breast_cancer()
df
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.33)

# Create a random forest classifier
rf = RandomForestClassifier(n_estimators=100, max_depth=5)

# Train the model on the training data
rf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
