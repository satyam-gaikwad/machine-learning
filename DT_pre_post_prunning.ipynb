```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

x_tr,x_te,y_tr,y_te = train_test_split(iris.data,iris.target)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## postprunning
treeM = DecisionTreeClassifier(max_depth=2)
```

```python
treeM.fit(x_tr,y_tr)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeM,filled=True)
```

```python
y_pred = treeM.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,y_te))
```

## Pre-prunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeM = DecisionTreeClassifier()
```

```python
parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
cv = GridSearchCV(treeM,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(x_tr,y_tr)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_pred,y_te))
```
